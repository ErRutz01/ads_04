# Building an predictive model?

Die Unterscheidung bezieht sich letztlich nur auf den Bereich des Supervised Learning.

- instance-based learning
- model-based learning

## Vorarbeit

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc("axes.spines", top=False, right=False)
plt.rcParams["figure.figsize"] = [10, 6]
plt.rcParams["figure.dpi"] = 100
```

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

```python
X = [[x] for x in tips["total_bill"]]
y = list(tips["tip"])
x_p = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
X_predict = [[x] for x in x_p]
x_lim = [0, 55]
y_lim = [0, 11]
```

```python
# Neue Fälle
x = sns.scatterplot(x="total_bill", y="tip", data=tips, color="lightgrey")
for i in x_p:
    x.axvline(i, color="magenta", linestyle=":")
x.set_xlim(x_lim)
x.set_ylim(y_lim)
```

## Instance-based Learning

Für jeden neuen Fall werden möglichst ähnliche Fälle (mit bekannten Labels) identifiziert. Dazu braucht es Distanzmaße um "Ähnlichkeit" messen zu können. Anschließend kann basierend auf den identifizierten Fällen eine Prognose getroffen werden. Dieses Vorgehen kann für einzelne Vorhersagen eine schnelle Lösung bilden, ist jedoch verhältnismäßig aufwendig, wenn viele neue Fälle vorhergesagt werden sollen.

```python
n_neighbors = 4
```

```python
from sklearn.neighbors import KNeighborsRegressor

model_01 = KNeighborsRegressor(n_neighbors=n_neighbors).fit(X, y)
results_01 = model_01.predict(X_predict)
```

```python
x = sns.scatterplot(x="total_bill", y="tip", data=tips, color="lightgrey")
x.scatter(x_p, results_01, color="red")
x.set_xlim(x_lim)
x.set_ylim(y_lim)
```

**Zur Erklärung:**

Jeder Wert wird durch die vier naheliegendsten Werte erklärt bzw. vorhergesagt.

```python
temp = tips.loc[(tips["total_bill"] > 42) & (tips["total_bill"] < 48.2)]
x = sns.scatterplot(x="total_bill", y="tip", data=tips, color="lightgrey")
x.scatter(x_p[8], results_01[8], color="red")
x.scatter(temp["total_bill"], temp["tip"], color="grey")
x.set_xlim(x_lim)
x.set_ylim(y_lim)
```

```python
tips.loc[(tips["total_bill"] > 42) & (tips["total_bill"] < 48.2)]
```

## Model-based Learning

Hier wird "generalisiert", es wird also ein abstraktes Modell entwickelt, welches anschließend sehr schnelle Vorhersagen ermöglicht.

```python
from sklearn.linear_model import LinearRegression

model_02 = LinearRegression().fit(X, y)
results_02 = model_02.predict(X_predict)
results_02_extremes = model_02.predict([[0], [60]])
```

```python
x = sns.scatterplot(x="total_bill", y="tip", data=tips, color="lightgrey")
x.scatter(x_p, results_02, color="blue")
x.set_xlim(x_lim)
x.set_ylim(y_lim)
```

**Zur Erklärung:**

Die Werte werden durch ein abstraktes Modell, in diesem Fall eine gerade Linie, vorhergesagt.

```python
x = sns.scatterplot(x="total_bill", y="tip", data=tips, color="lightgrey")
x.plot([0, 60], results_02_extremes, color="blue", linestyle="--")
x.set_xlim(x_lim)
x.set_ylim(y_lim)
```

## Comparison

```python
x = sns.scatterplot(x="total_bill", y="tip", data=tips, color="lightgrey")
x.plot([0, 60], results_02_extremes, color="blue", linestyle="--")
x.scatter(x_p, results_01, color="red")
x.set_xlim(x_lim)
x.set_ylim(y_lim)
```
