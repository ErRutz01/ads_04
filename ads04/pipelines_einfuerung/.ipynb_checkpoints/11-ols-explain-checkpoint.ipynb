# OLS-Regression (schließende Statistik)

https://www.statsmodels.org/dev/example_formulas.html

## Prepare

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc("axes.spines", top=False, right=False)
plt.rcParams["figure.figsize"] = [10, 6]
plt.rcParams["figure.dpi"] = 100
```

## Import

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

```python
df.shape
```

## Inspect

```python
sns.pairplot(data=df)  # , hue="origin")
```

```python
fig, ax = plt.subplots(2, 2, figsize=[10, 10])
sns.scatterplot(x="horsepower", y="mpg", data=df, alpha=0.25, ax=ax[0, 0])
sns.scatterplot(x="weight", y="mpg", data=df, alpha=0.25, ax=ax[0, 1])
sns.scatterplot(x="acceleration", y="mpg", data=df, alpha=0.25, ax=ax[1, 0])
sns.scatterplot(x="displacement", y="mpg", data=df, alpha=0.25, ax=ax[1, 1])
```

```python
fig, ax = plt.subplots(1, 3, figsize=[15, 5])
sns.boxplot(x="cylinders", y="mpg", data=df, color="lightblue", ax=ax[0])
sns.boxplot(x="model_year", y="mpg", data=df, color="lightblue", ax=ax[1])
sns.boxplot(x="origin", y="mpg", data=df, color="lightblue", ax=ax[2])
```

```python
corr = df.corr().round(2)
sns.heatmap(corr, vmin=-1, vmax=1, cmap="RdBu", annot=True)
```

## Analysis

```python
model = smf.ols(formula="mpg ~ weight", data=df)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula="mpg ~ displacement", data=df)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula="mpg ~ displacement + cylinders + horsepower + weight", data=df)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula="mpg ~ origin", data=df)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula="mpg ~ C(origin, Treatment('usa'))", data=df)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(
    formula="mpg ~ weight + model_year + C(origin, Treatment('usa'))", data=df
)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(
    formula="mpg ~ weight + np.square(weight) + model_year + C(origin, Treatment('usa'))",
    data=df,
)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula="mpg ~ weight + np.square(weight) + model_year", data=df)
results = model.fit()
print(results.summary())
```

## Review

```python
demo = pd.DataFrame(dict(weight=[(i * 100) + 1000 for i in range(50)]))
```

```python
model1 = smf.ols(formula="mpg ~ weight", data=df).fit()
model2 = smf.ols(formula="mpg ~ weight + np.square(weight)", data=df).fit()
predict1 = model1.predict(demo)
predict2 = model2.predict(demo)
```

```python
xlim = (df["weight"].min() - 500, df["weight"].max() + 500)
ylim = (0, df["mpg"].max() + 5)
fix, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(x="weight", y="mpg", data=df, color="black", alpha=0.2, ax=ax[0])
sns.scatterplot(x="weight", y="mpg", data=df, color="black", alpha=0.2, ax=ax[1])
ax[0].plot(demo["weight"], predict1, color="blue")
ax[1].plot(demo["weight"], predict2, color="blue")
for i in [0, 1]:
    ax[i].set_xlim(xlim)
    ax[i].set_ylim(ylim)
ax[0].set_title("Lineare Funktion")
ax[1].set_title("Quadratisches Polynom")
plt.tight_layout()
```
