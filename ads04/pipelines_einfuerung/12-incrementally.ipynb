# Can incrementally learn?

Was wir uns anschauen wollen:

- Batch oder offline learning
- Online learning

```python
a = [1, 3, 2, 1, 2, 3, 4, 2, 4, 2, 5, 3, 5, 3, 4, 6, 3, 5]
b = [4, 5, 6, 5, 6, 3, 4, 3, 5, 6, 7, 4, 5, 4, 6, 5]
c = [8, 7]
x = a + b + c
```

```python
def mean(x):
    return sum(x) / len(x)
```

```python
mean([1, 2, 3, 4])
```

```python
def median(x):
    r = sorted(x)
    i = round(len(x) / 2 - 0.5)
    return r[i]
```

```python
median([1, 3, 2])
```

```python
median([1, 4, 3, 2])
```

```python
def print_results(x):
    print("Length:", len(x))
    print("Mean:  ", mean(x))
    print("Median:", median(x))
```

## Batch Learning

**Batch Learning** (offline): Das Modell wird mit allen zur Verfüfung stehenden Daten trainiert, was resourcen-intensiv ist und daher meist offline durchgeführt wird. Allerdings kann dieser Prozess in vielen Fällen gut automatisiert werden, weswegen es dann kein großes Problem ist.

```python
print_results(x)
```

## Online Learning

**Online Learning**: Das Modell wird inkrementell entwickelt, wobei entweder einzelne Fälle oder kleinere Gruppen (_mini batches_) nacheinander einbezogen werden. Das System lernt also _on the fly_. Dies ist sowohl interesant, wenn kontinuierlich neue Daten entstehen, aber auch wenn die Daten zu groß sind um auf einen Schlag als _batch_ verarbeitet zu werden. Jedoch sind nicht alle Algorithmen direkt für das Online Learning geeignet und es kann zu Problemen kommen, wenn die Datenqualität über die Zeit nachlässt (z.B. wegen eines kaputten Sensors).

```python
print_results(a)
```

```python
print_results(b)
```

```python
print_results(c)
```

### Länge "online" lernen

```python
len_all = len(a) + len(b) + len(c)
len_all
```

### Mittelwert "online" lernen

```python
means = [mean(a), mean(b), mean(c)]
means
```

```python
mean_wrong = mean(means)
mean_wrong
```

```python
mean_all = (mean(a) * len(a) + mean(b) * len(b) + mean(c) * len(c)) / len_all
mean_all
```

### Median "online" lernen geht nicht so einfach

```python
medians = [median(a), median(b), median(c)]
medians
```

```python
median(medians)
```

### Offline Modell

```python
print_results(x)
```
