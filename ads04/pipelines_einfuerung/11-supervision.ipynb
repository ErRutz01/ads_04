## Supervised learning

**Supervised Learning:** Die Trainingsdaten beinhalten eine Lösung (_label_). Dabei werden vor allem Regressions- (numerisches Label) und Klassifikationsaufgaben (binäres oder ordinales Label) unterschieden.


```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

## Exkurs: Theoriegeleites Lernen

```python
temp = tips.copy()
```

```python
temp["generous"] = (temp["tip"] > temp["total_bill"] * 0.15) & (temp["tip"] > 2)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="generous", data=temp)
```

## Unsupervised learning

**Unsupervised Learning:** Es stehen keine Label zur Verfügung, das System versucht selbständig Muster zu finden. Dabei können insbesondere drei Gruppen von Methoden unterschieden werden: in der Cluster-Analyse werden Entitäten (Zeilen) gruppiert, bei der Anomaly-Detektion werden insbesondere Ausreißer identifiziert und in der Dimensionsreduktion werden Attribute (Spalten) zusammenfassen und so die Größe des Datensatzes reduziert.

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2, random_state=0)
results = model.fit(pd.get_dummies(tips))
tips["cluster"] = results.labels_
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="cluster", data=tips)
```

## Reinforcement learning

**Reinforcement Learning:** Diese Algorithmen verfolgen einen grundsätzlich anderen Ansatz. Das System beobachtet seine Umwelt und trifft eigenständig Entscheidungen, welche anschließend belohnt oder bestraft werden. Durch _try and error_ lernt das System erfolgsversprechende Lösugnsstrategien zu identifizieren.
