# 5 Feature engineering

```python
import os
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("ticks")
plt.rc("axes.spines", top=False, right=False)
plt.rcParams["figure.figsize"] = [10, 6]
plt.rcParams["figure.dpi"] = 100
plt.rcParams["axes.grid"] = False
plt.rcParams["grid.color"] = "white"
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import (
    StandardScaler,
    Normalizer,
    QuantileTransformer,
    MinMaxScaler,
    KBinsDiscretizer,
    PolynomialFeatures,
    FunctionTransformer,
)
from sklearn.impute import SimpleImputer
```

```python
TITANIC_PATH = os.path.join("..", "data", "titanic.csv")
```

## Import

```python
titanic_df = pd.read_csv(TITANIC_PATH)
```

## Allgemein

## Basics in Scikit-Learn

 Gliederung aktuell nach: https://scikit-learn.org/stable/modules/preprocessing.html

```python
age = titanic_df.loc[titanic_df["age"].notnull(), ["age"]]
age = age.dropna()
age = age.reset_index(drop=True)
age_input = age.copy()
age
```

### Standardization, or mean removal and variance scaling

```python
age["standard"] = StandardScaler().fit_transform(age_input)
age["min_max"] = MinMaxScaler().fit_transform(age_input)
```

```python
age
```

### Non-linear transformation

```python
age["quantile"] = QuantileTransformer(n_quantiles=100).fit_transform(age_input)
```

### Interlude: visualize results

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
sns.histplot(age["age"], ax=ax[0, 0], color="lightgrey")
sns.histplot(age["standard"], ax=ax[0, 1], color="lightgrey")
sns.histplot(age["quantile"], ax=ax[1, 1], color="lightgrey")
sns.histplot(age["min_max"], ax=ax[1, 0], color="lightgrey")
for i in range(2):
    for j in range(2):
        ax[i, j].tick_params(axis="x", colors="red")
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(x="age", y="min_max", data=age, ax=ax[0])
sns.scatterplot(x="age", y="quantile", data=age, ax=ax[1])
plt.tight_layout()
```

### Normalization

```python
X = [[1.0, -1.0, 2.0], [2.0, 0.0, 0.0], [5.0, -20.0, 10.0], [0.0, 1.0, -1.0]]
```

```python
Normalizer(norm="max").fit_transform(X)
```

### Encoding categorical features

```python
embarked = titanic_df.loc[:, ["embarked"]]
embarked = embarked.dropna()
embarked = embarked.reset_index(drop=True)
embarked_input = embarked.copy()
```

```python
embarked["numeric"] = OrdinalEncoder().fit_transform(embarked_input)
```

```python
one_hot = OneHotEncoder().fit_transform(embarked_input).toarray()
one_hot = pd.DataFrame(one_hot)
embarked = embarked.merge(one_hot, right_index=True, left_index=True)
```

```python
embarked.drop_duplicates()
```

### Discretization

```python
age["bins"] = KBinsDiscretizer(encode="ordinal").fit_transform(age_input)
```

```python
age
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 5))
sns.histplot(x="age", data=age, ax=ax[0])
sns.scatterplot(x="age", y="bins", data=age, ax=ax[1])
plt.tight_layout()
```

### Imputation of missing values

https://scikit-learn.org/stable/modules/impute.html

```python
temp = titanic_df.copy()
temp["age_imputed"] = SimpleImputer().fit_transform(temp.loc[:, ["age"]])
```

```python
temp["age"].mean()
```

```python
temp["age_imputed"].mean()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(temp["age"], ax=ax[0,])
sns.histplot(temp["age_imputed"], ax=ax[1])
ax[0].set_title("Original")
ax[1].set_title("Imputed")
plt.tight_layout()
```

### Generating polynomial features

```python
X = [[0, 1], [2, 3], [4, 5]]
```

$(x_1, x_2) \rightarrow (1, x_1, x_2, x_1^2, x_1 x_2, x_2^2)$

```python
PolynomialFeatures().fit_transform(X)
```

## Custom Transformers

```python
temp = titanic_df.loc[titanic_df["fare"] > 0].copy()
```

```python
temp["log_fare"] = np.log(temp["fare"])
```

```python
temp["quant_fare"] = QuantileTransformer(n_quantiles=100).fit_transform(
    temp.loc[:, ["fare"]]
)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(12, 8))
sns.histplot(temp["fare"], ax=ax[0, 0])
sns.scatterplot(x="fare", y="fare", data=temp, ax=ax[1, 0])
sns.histplot(temp["log_fare"], ax=ax[0, 1])
sns.scatterplot(x="fare", y="log_fare", data=temp, ax=ax[1, 1])
sns.histplot(temp["quant_fare"], ax=ax[0, 2])
sns.scatterplot(x="fare", y="quant_fare", data=temp, ax=ax[1, 2])
ax[0, 0].set_title("Ausgangsvariable")
ax[0, 1].set_title("Nach Log-Transformation")
ax[0, 2].set_title("QuantileTransformer")
plt.tight_layout()
```

```python
temp.corr()[["fare", "log_fare", "quant_fare"]].round(4)
```

```python
transformer = FunctionTransformer(np.log, validate=True)
```

```python
transformer.fit_transform(titanic_df.loc[:, ["fare"]])
```
