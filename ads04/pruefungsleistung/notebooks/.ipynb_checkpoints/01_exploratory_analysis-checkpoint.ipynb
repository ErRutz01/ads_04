# Explorative Datenanalyse

**Ziel des Notebooks**:

Wir nutzen dieses Notebook für die Explorative Datenanalyse (EDA), um die Datenstruktur, Verteilungen, Ausreißer und das Klassenungleichgewicht der Daten zu verstehen. Die gewonnenen Erkenntnisse dienen als Entscheidungsgrundlage dafür, ob ein Klassifikations- oder ein Regressionsansatz für unser Zielvorhaben – die Vorhersage von Retouren – besser geeignet ist.

Dafür untersuchen wir zwei unterschiedliche Zielvariablen:

- Klassifikationsansatz: Die binäre Variable `num_returned`, um das Eintreffen einer Retoure als diskretes Ereignis vorherzusagen.

- Regressionsansatz: Die neu definierte, stetige Variable `proxy_returnrate`, um die relative Retourenhäufigkeit auf aggregierter Produktebene abzubilden.

Im Verlauf dieses Notebooks analysieren wir die Zusammenhänge der potenziellen Features mit beiden Zielvariablen, um die tragfähigste Modellierungsstrategie für den weiteren Verlauf zu bestimmen.

**NOTE:**

---

In den Verteilungen beider Zielvariablen wird ein ausgeprägtes Ungleichgewicht sichtbar, welches für die spätere Modellierung eine zentrale methodische Herausforderung darstellt:


`num_returned`

-  Hier dominieren die negativen Fälle (Klasse 0: nicht retourniert) als absolute Mehrheitsklasse. Für die Klassifikationsmodelle besteht hierbei die Gefahr der sogenannten „Accuracy-Falle“. Ein Modell könnte eine vermeintlich hohe Gesamtgenauigkeit erzielen, indem es pauschal die Mehrheitsklasse vorhersagt, ohne tatsächliche Retouren (Minderheitsklasse) zu erkennen. Die Evaluation muss daher primär über Recall und Präzision stattfinden.


`proxy_returnrate`

- Die Verteilung ist extrem „zero-inflated“ und weist eine starke Rechtsschiefe auf. Für klassische, lineare Regressionsmodelle ist diese Struktur hochgradig problematisch, da Standard-Verfahren homogene Varianz und Normalverteilung der Fehlerterme voraussetzen.

### Gruppenunterschiede

#### ANOVA

**NOTE:**

---

- Im Folgenden untersuchen wir die Gruppenunterschiede zwischen den kategorialen Variablen.

- Als Werkzeug nutzen wir ANOVA. Wir vergleichen die Mittelwerte zwischen den Gruppen gegen den Target`proxy_returnrate`.

- Variablen mit signifikanten Gruppenunterschieden, hergeleitet aus den P-Werten, bieten die stärkste Erklärungskraft und Grundlage für Klassifikationsmodelle.

---

```python
def anova(df, var):
    
    kategorien = df[var].unique()
    
    gruppen_liste = [df[df[var] == kat]["proxy_returnrate"] for kat in kategorien]
    
    f_statistic, p_value = f_oneway(*gruppen_liste)
    
    return f_statistic, p_value
```

```python
for m in cat_var:
    merkmal = m

    f_statistic, p_value = anova(merged_superstore_df, merkmal)

    print(f"{merkmal}: {f_statistic}")
    print(f"{merkmal}: {p_value}")
```

#### Chi-Quadrat-Test

Während eine ANOVA Gruppenunterschiede bei stetigen Variablen untersucht, nutzen wir hier den Chi-Quadrat-Test, da wir Gruppenunterschiede für das kategoriale Target `num_returned` analysieren.

```python
for col in cat_var:

    contingency_table = pd.crosstab(merged_superstore_df[col], merged_superstore_df["num_returned"])

    chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

    
    print(f"Chi-Quadrat-Test für {col}:")
    print(f"Chi2-Statistik: {chi2_stat:.4f}")
    print(f"p-Value:        {p_val:.4e}")
    print(
        f"Ergebnis:       {'SIGNIFIKANTER UNTERSCHIED (Abhängig)' if p_val < 0.05 else 'Kein signifikanter Unterschied (Unabhängig)'}"
    )
    print("-" * 40)
```

#### Mann-Whitney-U-Test

Mit diesem Test untersuchen wir den Einfluss unserer numerischen Metriken (wie Sales oder Profit) auf das Retourenverhalten.

```python
for col in num_var:
    
    group_0 = merged_superstore_df[merged_superstore_df["num_returned"] == 0][col]
    group_1 = merged_superstore_df[merged_superstore_df["num_returned"] == 1][col]

    stat, p_val = mannwhitneyu(group_0, group_1)

    print(f"Mann-Whitney-U-Test für {col}:")
    print(f"p-Value: {p_val:.4e}")
    print(
        f"Ergebnis: {'Signifikant unterschiedliche Verteilungen' if p_val < 0.05 else 'Kein Unterschied'}\n"
    )
```

#### Transform `Discount` to category Variable

**Note**:

---
Im Abschnitt Verteilung analysieren fällt auf, dass die Variable Discount zwar numerisch erfasst ist, in der Praxis aber eine rein diskrete Struktur besitzt (Rabatte werden im Online-Shop in festen Prozentsätzen wie 0%, 10% oder 20% vergeben).

Da es sich hierbei nicht um eine kontinuierlich frei fließende Metrik wie bei Sales oder Profit handelt, ist eine Einteilung in feste Rabattklassen (Binning) fachlich sinnvoller als eine rein numerische Skalierung.

Wir konvertieren Discount daher zu einer kategorialen Variable (cat_discount) mit folgenden Ausprägungen:

`no_discount`

`low_discount`

`medium_discount`

`high_discount`

Andere schiefe Variablen (wie Sales oder Profit) belassen wir in ihrer stetigen Form und passen sie stattdessen über den StandardScaler in der Pipeline an, da dort jeder einzelne Cent-Betrag eine Rolle spielt.

---

```python
def set_class(d_value):
    if d_value == 0:
        return "no_discount"
    elif (d_value > 0) & (d_value <= 0.2):
        return "low_discount"
    elif (d_value >= 0.2) & (d_value <= 0.5):
        return "medium_discount"
    else:
        return "high_discount"
```

```python
merged_superstore_df["cat_discount"] = merged_superstore_df["Discount"].apply(set_class)
```
