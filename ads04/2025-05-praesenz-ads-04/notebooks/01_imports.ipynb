# Import

````python
import os
import json
from lxml import etree
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
````

````python
INPUT_DIR = os.path.join("..", "input")
TEMP_DIR = os.path.join("..", "temp")
FLIGHTS_CSV = os.path.join(INPUT_DIR, "flights.csv")
AIRPORTS_CSV = os.path.join(INPUT_DIR, "airports.csv")
PLANES_CSV = os.path.join(INPUT_DIR, "planes.csv")
AIRLINES_XML = os.path.join(INPUT_DIR, "airlines.xml")
WEATHER_JSON = os.path.join(INPUT_DIR, "weather.json")
TRAIN_PICKLE = os.path.join(TEMP_DIR, "train.pickle")
TEST_PICKLE = os.path.join(TEMP_DIR, "test.pickle")
````

## Import flights.csv

````python
flights_df = pd.read_csv(FLIGHTS_CSV)
````

````python
flights_df.head()
````

## Import airports.csv

````python
airports_df = pd.read_csv(AIRPORTS_CSV)
````

````python
airports_df.head()
````

## Import planes.csv

````python
planes_df = pd.read_csv(PLANES_CSV)
````

````python
planes_df.head()
````

## Import weather.json

````python
with open(WEATHER_JSON, "r", encoding="utf-8") as f:
    weather_json = json.load(f)
````

````python
weather_df = pd.DataFrame(weather_json)
````

````python
weather_df.head()
````

## Import airlines.xml

Struktur der Datei:

```xml
<airlines>
  <airline carrier="9E">Endeavor Air Inc.</airline>
  <airline carrier="AA">American Airlines Inc.</airline>
  <airline carrier="AS">Alaska Airlines Inc.</airline>
</airlines>
```

````python
airlines_tree = etree.parse(AIRLINES_XML)
````

````python
airlines_root = airlines_tree.getroot()
````

````python
rows = []
````

````python
for a in airlines_root.findall("airline"):
    row = a.attrib
    row["name"] = a.text
    rows.append(row)
````

````python
rows
````

````python
airlines_df = pd.DataFrame(rows)
````

````python
airlines_df
````

## Merge 1: flights + planes

````python
flights_df.shape
````

````python
planes_df.shape
````

````python
merged_df = pd.merge(
    flights_df.drop("Unnamed: 0", axis=1),
    planes_df.drop("Unnamed: 0", axis=1),
    on="tailnum",
    suffixes=["", "_plane"],
    how="left",
)
````

````python
merged_df.shape
````

````python
merged_df.columns
````

````python
merged_df.head()
````

## Merge 2: + weather

````python
merged_df.shape
````

````python
weather_df.shape
````

````python
merged_df = pd.merge(
    merged_df,
    weather_df,
    on=["origin", "time_hour"],
    suffixes=["", "_weather"],
    how="left",
)
````

````python
merged_df.shape
````

## Train-Test-Split

````python
train_df, test_df = train_test_split(merged_df, train_size=.8, random_state=42)
````

````python
merged_df.shape
````

````python
train_df.shape
````

````python
test_df.shape
````

## Export

````python
train_df.to_pickle(TRAIN_PICKLE)
````

````python
test_df.to_pickle(TEST_PICKLE)
````
